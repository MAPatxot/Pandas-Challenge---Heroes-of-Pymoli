## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Bins - Activity 4 > TED Talks
# bins -- slice data
# pd.cut() -- (columns, bins, labels)
# Total Count = 576 -- existing players -- like gender, create one for age

age_players = purchase_data.loc[:,["Age","SN"]]
age_players = age_players.drop_duplicates()
age_existing_players = age_players["Age"].value_counts()
# pd.DataFrame(age_total)

bins_age = [0, 9, 14, 19, 24, 29, 34, 39, 100]

labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_data = pd.cut(age_players["Age"], bins_age, labels=labels).value_counts()
# pd.DataFrame(age_data)

age_percent = age_data / total_players * 100
# pd.DataFrame(age_percent)

age_demographics = pd.DataFrame({"TOTAL COUNT": age_data,
                              "PERCENTAGE OF PLAYERS": age_percent})
age_demographics["PERCENTAGE OF PLAYERS"] = age_demographics["PERCENTAGE OF PLAYERS"].map("{:.2f}".format)

# age_demographics >> labels are not in order, need to fix!
# sort_index -- Sort object by labels (along an axis)

age_demographics.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# bin the purchase_data

purchase_data["Ranges of Age"] = pd.cut(purchase_data["Age"], bins_age, labels=labels)
# purchase_data.head()

age_analysis_total = purchase_data.groupby(["Ranges of Age"]).count()["Price"].rename("PURCHASE COUNT")

age_average_price = purchase_data.groupby(["Ranges of Age"]).mean()["Price"].rename("AVERAGE PURCHASE PRICE")

age_total_purchase = purchase_data.groupby(["Ranges of Age"]).sum()["Price"].rename("TOTAL PURCHASE VALUE")

age_totalpurchase_perperson = age_total_purchase / age_data
# I didn't need to create an age_total, like I did for gender_total

age_purchasing_analysis = pd.DataFrame({"PURCHASE COUNT": age_analysis_total,
                              "AVERAGE PURCHASE PRICE": age_average_price,
                              "TOTAL PURCHASE VALUE": age_total_purchase,
                              "AVG TOTAL PURCHASE PER PERSON": age_totalpurchase_perperson})

age_purchasing_analysis["AVERAGE PURCHASE PRICE"] = age_purchasing_analysis["AVERAGE PURCHASE PRICE"].map("${:.2f}".format)
age_purchasing_analysis["TOTAL PURCHASE VALUE"] = age_purchasing_analysis["TOTAL PURCHASE VALUE"].map("${:.2f}".format)
age_purchasing_analysis["AVG TOTAL PURCHASE PER PERSON"] = age_purchasing_analysis["AVG TOTAL PURCHASE PER PERSON"].map("${:.2f}".format)
                                                                                                                    

age_purchasing_analysis

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# item_players = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
# item_players = item_players.drop_duplicates()
# item_existing_players = item_players["Item Name"].value_counts()
# pd.DataFrame(item_existing_players)

item_id = purchase_data.groupby("Item ID").sum().reset_index()
item_name_sum = purchase_data.groupby("Item Name").sum().reset_index()
item_name_count = purchase_data.groupby("Item Name").count().reset_index()

item_id_sum = pd.merge(item_id, item_name_sum, on="Price")
# item_id_sum.head()

item_id_count = pd.merge(item_name_count, item_id_sum, on="Item Name")
# item_id_count.head()

item_id_count["Gender"] = (item_id_count["Price_y"]/item_id_count["Item ID"]).round(2)

item_id_count_rename = item_id_count.rename(columns={"Age": "PURCHASE COUNT", "Gender": "ITEM PRICE", "Item ID": "null", "Price_y": "TOTAL PURCHASE VALUE", "Item ID_y": "Item ID"})
# item_id_count_rename

clean_item = item_id_count_rename[["Item ID", "Item Name", "PURCHASE COUNT", "ITEM PRICE", "TOTAL PURCHASE VALUE"]]
# clean_item

popular_item = clean_item.set_index(["Item ID", "Item Name"])
# popular_item

popular_item_total = popular_item.sort_values("PURCHASE COUNT", ascending=False).head(5)
# popular_item_total

popular_item_total.style.format({"ITEM PRICE": "${:.2f}", "TOTAL PURCHASE VALUE": "${:.2f}"})

```
